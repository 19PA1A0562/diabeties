```
data['gender'] = data.gender.map(dict(M=1, F=0))
```

```
from sklearn.linear_model import LogisticRegression
```

```
x = data.loc[:,"encounter_id":"solid_tumor_with_metastasis"]
y = data['diabetes_mellitus']
```

```
model = LogisticRegression(max_iter=2000)
model.fit(x,y)
```

```
y_train_predict = model.predict(x)
```

```
print(pd.DataFrame({"Actual":data["diabetes_mellitus"],"Predicted":y_train_predict}))
```

```
from sklearn.metrics import roc_auc_score
```

```
roc_auc_score(y,y_train_predict)
```
